### Converting the age value to integer/numeric

```python
patients["age"] = pd.to_numeric(patients["age"], errors = "coerce") #Replace anything that cannot be converted to a number with NaN(Not a Number). Invalid age become NaN
```

```python
#Replace missing age values
patients["age"] = patients["age"].fillna(
    patients["age"].median()
    )
```

```python
patients["age"]  # Here, we are still getting float value and negative values. we will convert the age values to integer and replace negative valiue with the median value.
```

```python
# Replace negative ages with missing  values
patients.loc[patients["age"]< 0, "age"] = pd.NA  # what this line of code does is that it replace any negative age values with amissing value (NaN). This is done because negative age are not valid and should be treated as missing date. By replacing them with NaN, we can then handle them appropriately, such as filling them with the median age value later on.

# Fill missing with the median()
patients["age"] = patients["age"].fillna(patients["age"].median())


# Round age
patients["age"] = patients["age"].round()

# convert to integers
patients["age"] = patients["age"].astype(int)
```

#### Remove spaces in the phone_number column

```python
patients["phone_number"] = patients["phone_number"].str.replace(" ", "", regex=False) # this means that any spaces in the phone_number column will be removed. The regex=False argument indicates that we are not using regular expressions for this replacement, but rather a simple string replacement. This is done to ensure that the phone numbers are in a consistent format without any spaces.
```

#### Remove Hyphens in the phone_number column

```python
patients["phone_number"] = patients["phone_number"].str.replace(" ", "", regex=False) # this means that any spaces in the phone_number column will be removed. The regex=False argument indicates that we are not using regular expressions for this replacement, but rather a simple string replacement. This is done to ensure that the phone numbers are in a consistent format without any spaces.
```

#### Convert +234 numbers to Nigerian format

```python
patients["phone_number"] = patients["phone_number"].str.replace(
    r"^\+234",
    "0",
    regex=True
)
```

```python
patients["phone_number"]
```

### Addding the missing leading zero

```python
patients["phone_number"] = patients["phone_number"].apply(
    lambda x: "0" + x
    if isinstance(x, str) and x.isdigit() and len(x) == 10
    else x
)
```

```python
patients["phone_number"]
```

```python
patients["phone_number"] = patients["phone_number"].replace({
    "not_available": "Unknown"
})
```

```python
patients["phone_number"]
```

```python
# Replace invalid phone numbers
# A Nigerian phone number should contain 11 digits.
#Anything shorter or longer is invalid. 

patients["phone_number"] = patients["phone_number"].apply(
    lambda x: "Unknown"
    if isinstance(x, str)
    and x != "Unknown"
    and (not x.isdigit() or len(x) != 11)
    else x
)

"""
This will automatically convert:
12345 into Unknown

while leaving 08026951234 unchanged.
"""
```

##### Clean the appointment_id missing rows/values

Since appointmnent_id is the primary key of the appointmnent table, you should not fill missing values with "Unknown" or leave them blank.


A primary key must satisfy three rules:



.   Unique

.   Not Null

.   Stable (should not chnage)


.



```python
# Find the largest appointment_id
# Get the highest appointment number
max_id = (
    appointments["appointment_id"]
    .dropna()
    .str.replace("A", "", regex=False)
    .astype(int)
    .max()
)

print(max_id)
```

```python
# Generate new appointment IDs for missing values 
# Find rows with missing IDs
missing_rows = appointments["appointment_id"].isna()

# Create new IDs
new_ids = [
    f"A{num:06d}"
    for num in range(max_id + 1, max_id + 1 + missing_rows.sum())
]

# Assign them
appointments.loc[missing_rows, "appointment_id"] = new_ids
```

```python
# Verify that there are no more missing IDs
appointments["appointment_id"].isnull().sum()
```

```python
appointments.isnull().sum()
```
